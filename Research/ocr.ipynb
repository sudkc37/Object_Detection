```python
import easyocr
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#instensiate the OCR
img_reader = easyocr.Reader(['en'], gpu=False)
```

```python
#Extract the text from the image
extract_txt = img_reader.readtext('/Users/sudipkhadka/Desktop/Object_Detection/images/New_Jersey_license_plate_2014.jpg')
```

```python
extract_txt
```

```python
##plot the text back to image
left_top_cord = tuple(extract_txt[0][0][0])
right_top_cord = tuple(extract_txt[0][0][2])
text = extract_txt[0][1]
```

```python
left_top_cord
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread('/Users/sudipkhadka/Desktop/Object_Detection/images/New_Jersey_license_plate_2014.jpg')
img = cv2.rectangle(img, left_top_cord, right_top_cord, (0,255,0), 3)
img = cv2.putText(img, text, left_top_cord, font, 2, (0,0,255), 2, cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
images = cv2.imread('/Users/sudipkhadka/Desktop/Object_Detection/images/images.png')
extract_txt = img_reader.readtext('/Users/sudipkhadka/Desktop/Object_Detection/images/images.png')
for texts in extract_txt:
    left_top_cord = tuple([int(val) for val in texts[0][0]])
    right_top_cord = tuple([int(val) for val in texts[0][2]])
    text = texts[1]
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Calculate bounding box width and height
    box_width = right_top_cord[0] - left_top_cord[0]
    box_height = right_top_cord[1] - left_top_cord[1]

    # Adjust font scale dynamically
    font_scale = min(box_width / (len(text) * 20), box_height / 30)
    font_scale = max(font_scale, 0.5)  # Ensure the font is not too small

    # Draw rectangle and put text
    img = cv2.rectangle(images, left_top_cord, right_top_cord, (0, 255, 0), 1)
    img = cv2.putText(images, text, left_top_cord, font, font_scale, (255, 255, 255), 1, cv2.LINE_AA)

# Display the image
plt.figure(figsize=(10, 10))
plt.imshow(images)
plt.axis("off")
plt.show()
```

```python
def text_extractor(path):

    extract_txt = img_reader.readtext(path)
    img = cv2.imread(path)
    img = img.copy()
    for texts in extract_txt:
        left_top_cord = tuple([int(val) for val in texts[0][0]])
        right_top_cord = tuple([int(val) for val in texts[0][2]])
        text = texts[1]
        font = cv2.FONT_HERSHEY_SIMPLEX
        img = cv2.rectangle(img, left_top_cord, right_top_cord, (0,255,0), 2)
        img = cv2.putText(img, text, left_top_cord, font, 2, (255,255,255), 1, cv2.LINE_AA)
    plt.figure(figsize=(10,10))
    plt.imshow(img)
    plt.show() 
```
