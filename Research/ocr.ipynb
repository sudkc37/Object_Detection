```python
import easyocr
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#instensiate the OCR
img_reader = easyocr.Reader(['en'], gpu=False)
```

```python
#Extract the text from the image
extract_txt = img_reader.readtext('/Users/sudipkhadka/Desktop/Object_Detection/images/New_Jersey_license_plate_2014.jpg')
```

```python
extract_txt
```

```python
##plot the text back to image
left_top_cord = tuple(extract_txt[0][0][0])
right_top_cord = tuple(extract_txt[0][0][2])
text = extract_txt[0][1]
```

```python
left_top_cord
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread('/Users/sudipkhadka/Desktop/Object_Detection/images/New_Jersey_license_plate_2014.jpg')
img = cv2.rectangle(img, left_top_cord, right_top_cord, (0,255,0), 3)
img = cv2.putText(img, text, left_top_cord, font, 2, (0,0,255), 2, cv2.LINE_AA)
plt.imshow(img)
plt.show()
```
